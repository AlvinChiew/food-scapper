## TH

```python
with open("../th_food_item_options_raw_proc.csv", "w", encoding="utf-8") as fw:
    with open("../th_food_item_options_raw.csv", "r", encoding="utf-8") as fr:
        
        data = fr.readlines()
        data_proc = []
        idx = -2
        last_food_id = ''
        for row in data:
            food_id = row.split('|')[0]
            if food_id != last_food_id:
                idx += 1
                last_food_id = food_id
            data_proc.append(str(idx) + '|' + row)
            
    fw.writelines(data_proc)
```

```python
df = pd.read_csv("../th_food_item_options_raw_proc.csv", encoding="utf8", delimiter="|")
df.columns = ['idx'] + list(df.columns[1:])
df['idx'] = df['idx'].astype(str) + df['food_id'].astype(str)

print(len(df['food_id'].drop_duplicates()))
print(df['attribute'].drop_duplicates().to_list())
df
```

```python
df = df.fillna('')
df['value'] = df['value1'].astype(str) + df['value2'].astype(str)
df = df[['idx', 'attribute', 'value']]
```

```python
df = df.pivot(index='idx', columns='attribute', values='value')
df.columns.name = None
df.index.name = None
```

```python
cols = ["english_name", "base_weight", "Energy, by calculation", "Total energy - by calculation", "Protein, total", "Fat, total", "Carbohydrate, available", "Carbohydrate, total", "Dietary fibre", "Sodium"]
df = df[cols]
num_cols = cols[2:]
num_cols
```

```python
df
```

```python
df_proc = df.copy()
for col in num_cols:
    df_proc[col] = df_proc[col].str.replace(',', '.')
    df_proc[col] = df_proc[col].str.extract(r'(\d+[,.]?\d?)').astype('float')
    df_proc[col] = df_proc[col].fillna(0)
df_proc
```

```python
# 'Energy, by calculation' + 'Total energy - by calculation'
# df[~df['Energy, by calculation'].isna() & ~df['Total energy - by calculation'].isna()]
# 'Carbohydrate, available' + 'Carbohydrate, total'
df[~df['Carbohydrate, available'].isna() & ~df['Carbohydrate, total'].isna()]
```

```python
df_proc['energy'] = df_proc['Energy, by calculation'].astype(float) + df_proc['Total energy - by calculation'].astype(float)
df_proc['carbs'] = df_proc['Carbohydrate, available'].astype(float) + df_proc['Carbohydrate, total'].astype(float)
cols = ["english_name", "base_weight", "energy", "Protein, total", "Fat, total", "carbs", "Dietary fibre", "Sodium"]
df_proc = df_proc[cols]
df_proc['portion_size'] = 100
df_proc['region'] = 'TH'
df_proc
```

```python
df_proc.columns = output_cols
df_proc
```

```python
df_proc.to_csv("../th_food_item_options.csv")
```
